```python
soc2 = soc.copy()
```

```python
soc2 = soc.rename(columns={'club' : 'Club','last_name': 'Last name', 'first_name': 'First Name', 'position' : 'Position', 
                           'base_salary' : 'Base Salary', 'guaranteed_compensation' : 'Guaranteed Compensation'})
```

# Who are the highest paid players in the sport.

```python
# Creates a variable to hold the 3 largest values in column 'Base Salary'
Top_3_Salaries = soc2.nlargest(3, ['Base Salary'])
```
