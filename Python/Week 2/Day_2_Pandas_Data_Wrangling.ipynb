## Dropping Columns
When looking through your dataset, you’ll often find many, many columns.  Most of them might not be necessary for your analysis – so it makes sense to drop some of these columns.

```python
# We're removing anything that is not essential in the data
# Creates a list of columns that you'd like to drop then pass it into the drop() function
# inpalce = True tells Python to remove those columns from our current variable, df
columns_to_be_dropped = ['Edition Statement','Corporate Author', 'Corporate Contributors', 'Former owner','Engraver']
```

```python
type(columns_to_be_dropped)
```

```python
# Now we have less columns (essentials)
# Axis=1 tells Python to look on the column axis
df = book_df.drop(columns_to_be_dropped, axis = 1)
df.head()
```

## Changing the index

```python
# First we check to see if our new index will be unique
df['Identifier'].is_unique
```

```python
# Then we set the new index
# Makes the Identifier column the identifier for the data
new_df = df.set_index('Identifier')
new_df.head()
```

# Class Group Work
Property Data Cleaning

```python
# After analyzing the original data we're adding the missing_values variable in order to get
# rid of non-standard missing values
```

```python
missing_values = ['n/a', 'na', '--']
prop = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/property.csv',
                   na_values = missing_values)
```

```python
prop.head()
```

## Exploratory Data Analysis

```python
prop.shape
```

```python
prop.info()
```

```python
prop.describe
```

```python
prop.columns
```

```python
prop2 = prop.copy()
```

```python
# Renamed a few columns
prop2 = prop.rename(columns={'ST_NUM' : 'Number','ST_NAME': 'Name', 'NUM_BEDROOMS': 'Num. of Bedrooms', 'NUM_BATH' : 'Num. of Baths'})
```

```python
# Filled in NaN value in the PID column
prop2['PID'].fillna(100005000.0, inplace = True)
prop2
```

## Unexpected Missing Values
Example – if a feature is suppose to be ‘string’ but a ‘numeric’ is there instead – it’s technically a missing value. Lets look at the OWN_NUM_BATH column.  In the 7th row there is a ‘HURLEY’ but it should be num.

```python
# Converts NaN to 0 in the Number column
# Replace all Nas in a certain column with one value:
prop2['Number'].fillna(0, inplace = True)
prop2
```

```python
# Replace certain values
prop2.loc[2, 'Number'] = 125
prop2.head()
```

```python
# Pandas function to search for incorrect data type and convert it
prop2['Num. of Baths'] = pd.to_numeric(prop2['Num. of Baths'], errors='coerce')
prop2
```

```python
 prop2.replace("12", np.nan, inplace=True) # Non-standard value in OWN_OCCUPIED
# prop2.replace("HURLEY", np.nan, inplace=True)
# prop2.replace("--", np.nan, inplace=True)
```

```python
# Replace using median value (a very common approach)
median = prop2['Num. of Bedrooms'].median()
prop2['Num. of Bedrooms'].fillna(median, inplace = True)
```
