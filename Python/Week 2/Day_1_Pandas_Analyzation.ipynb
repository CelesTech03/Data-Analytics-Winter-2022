```python
# Import data into our notebook
serv = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/311_Service_Requests_from_2019May.csv')
```

```python
serv.head()
```

```python
# Groupby, similar to what we’ve done in SQL, allows us to make summaries on certain columns.
# (['column you want to group'])['Unique Key']
serv.groupby(['Borough'])['Unique Key'].count()
```

```python
# In Python, multiple operations can be chained together using the dot method.
serv.groupby(['Borough'])['Unique Key'].count().sort_values(ascending = False)
```

```python
# Puts data into a bar plot
serv.groupby(['Borough'])['Unique Key'].count().sort_values(ascending = False).plot(kind = 'bar')
```

```python
serv.head()
```

```python
# Filtering Rows
serv[serv['Complaint Type'] == 'Noise']
serv[serv['Complaint Type'] == 'Noise'].count()
```

```python
# Python also allows us to do ‘fuzzy matching’ – which matches words that may be “less than 100% perfect”
serv[serv['Complaint Type'].str.contains('Noise')]
```

```python
serv.head()
```

```python
serv.groupby(['Borough'])['Unique Key'].count()
```

```python
# There are a few techniques to check for missing values.
# First, lets create a sample DF with some missing values to work with.
df = {'A': [1,2,np.nan], 'B':[3,np.nan, np.nan]}
df = pd.DataFrame(df)
```

```python
df.head()
```

```python
# is.null() is one way – however for large datasets it might not workout well.
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
# Lets focus on dropping missing values
df.dropna() # Drops all missing values
```

```python
# Drops NAs only from columns that contain NA values.  If a column doesn’t have any missing values, it wont be dropped.
df.dropna(axis=1)
```

```python
# Sets a threshold – NAs will only be dropped if that threshold is met.
df.dropna(thresh=2)
```
