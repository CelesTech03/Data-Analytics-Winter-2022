```python
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
data = pd.DataFrame(randn(5, 4), ['A', 'B', 'C', 'D', 'E'], 
                    ['W', 'X', 'Y', 'Z'])
```

#Selecting Rows

```python
data.head()
```

```python
# Drops row E from the DataFrame
data.drop('E', axis = 0)
```

```python
# loc uses the name of the element
data.loc['C']
```

```python
# iloc uses index to pull out the element
data.iloc[2] 
```

#Filtering

```python
# Prints where the values are true or false for greater than 0
data > 0
# nba['team']['players']['pts'] >  30
```

```python
# this line is saying show me all the values within the table 
# in which the expression is true; meaning, where all the values
# whiin the table are less than 0.
data < 0
```

```python
# This line translates to:
# let's hold the outcome of the expression 'data < 0' into a variable
booldf = data < 0
booldf
```

```python
# Only prints the data where all the values are True
# In this case all data less than 0
data[booldf]
```

```python
data[data < 0]
```

```python
# Only shows the values less than 0 in column W
# Notice it only shows row C
data[data['W'] < 0]
```

```python
# Only shows the values less than 0 in column W
# Notice it only shows row C
data[data['W'] > 0]
```
